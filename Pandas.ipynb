# Modify Values

# Pandas DataFraame /df

```python
nama_makanan =  pd.Series(['Nasi goreng', 'Nasi Ayam' , 'Mie Goreng'])
harga_makanan = pd.Series([15000,16000,14000])
rating_makanan = pd.Series([4.6,4.7,4.2])


print(nama_makanan, harga_makanan, rating_makanan)
```

```python
df = pd.DataFrame({
    'Nama Makanan' : nama_makanan,
    'Harga' : harga_makanan,
    'Rate' : rating_makanan
})

df
```
